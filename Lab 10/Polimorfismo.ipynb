# Pilares da OO: Polimorfismo

* Poli: muitas
* Morphos: formas

O polimorfismo permite tratar objetos de diferentes tipos como o mesmo tipo geral. Assumindo que esses diferentes tipos obedecem um mesmo padrão de comportamento, podemos escrever códigos mais gerais, capazes de trabalhar com quaisquer desses tipos.

<img src="https://drive.google.com/uc?id=1mKztPpEvd0KE-mFYruZ1J1j-BKzvhBDU" width=600>


Já vimos a expressão do polimorfismo no Python com a **sobrecarga de operadores**, onde redefinimos o comportamento de operadores aritméticos (+, -, *, /) e relacionais (>, <, ==, etc.), além do comportamento de funções como o print (usando o método mágico `__str__()`) e operadores de lista.

Podemos entender o polimorfismo com a sobrecarga de operadores da seguinte forma: ao ver uma expressão `a + b`, eu preciso saber qual o tipo de dados de `a` e de `b` para determinar o comportamento. A expressão escrita é a mesma, mas o comportamento vai variar. **Veja o exemplo a seguir com diferentes aplicações da soma**.




````python
# soma de dois inteiros é operação aritmética
a, b = 10, 20
#################
print(a+b) # soma
#################

# soma de duas strings é concatenação
a, b = 'A', 'B'
#################
print(a+b) # soma
#################

# soma de dois pontos pode ser definida como
# operação geométrica
class Ponto:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self, outro_ponto):
        return(self.x + outro_ponto.x,
               self.y + outro_ponto.y)

a = Ponto(3,4)
b = Ponto(5,6)
#################
print(a+b) # soma
#################
````

### Sobrescarga de métodos com Herança

No contexto da herança, o polimorfismo aparece quando reescrevemos um método da superclasse nas subclasses. Vejamos o exemplo a seguir onde modelamos mensagens enviadas e recebidas em um aplicativo que dá suporte a mensagens de Texto, Voz e Imagem.

*   Superclasse `Mensagem` representa o tipo mais comum, mensagens de texto, com o método `exibir` sendo apenas um print e um método `exibirTamanho` que imprime quantos bytes tem a mensagem.
*   `MensagemVoz` é uma subclasse que precisa de um tratamento especial para `exibir`, dar play no áudio.
*   `MensagemImagem` é outra subclasse que também trata de forma diferente o comportamento de `exibir` mensagem.



````python
import sys

class Mensagem:
    def __init__(self, conteudo):
        self.conteudo = conteudo

    def exibirTamanho(self):
        print(f'({sys.getsizeof(self.conteudo)} bytes)')

    def exibir(self):
        print(self.conteudo, end='')

class MensagemVoz(Mensagem):
    def __init__(self, conteudo):
        super().__init__(conteudo)

    def exibir(self):
        print(f'Dando play no arquivo de som {self.conteudo}', end='')

class MensagemImagem(Mensagem):
    def __init__(self, conteudo):
        super().__init__(conteudo)

    def exibir(self):
        print(f'Exibindo imagem {self.conteudo}', end='')
````

Agora pense numa classe `Conversa` que pode incluir todos os tipos de mensagens. A vantagem do polimorfismo é que a classe `Conversa` não precisa saber qual tipo específico de mensagem está armazenando. Ela simplesmente chama o método `exibir()`, e o tipo da variável vai determinar qual método será chamado! **Isso é polimorfismo!**

Se amanhã você precisar adicionar um novo tipo de mensagem, como `MensagemVideo`, basta criar a nova subclasse sobrescrevendo o método `exibir()` e a classe `Conversa` continuará funcionando sem sofrer nenhuma alteração.

Note que o método `exibirTamanho`, que também pode ser chamado pra qualquer classe na relação de herança, **não caracteriza polimorfismo** porque só existe um método implementado (o da superclasse) e é sempre ele a ser chamado.

````python
class Conversa:
    def __init__(self):
        self.mensagens = []

    def adicionar_mensagem(self, mensagem):
        self.mensagens.append(mensagem)

    def exibir_todas(self):
        for i, mensagem in enumerate(self.mensagens, 1):
            print(f'{i}. ', end='')
            mensagem.exibir() ### polimorfismo!
            print(' ', end='')
            mensagem.exibirTamanho() ### não é polimorfismo!

conversa = Conversa()
conversa.adicionar_mensagem(Mensagem("Olá, tudo bem?"))
conversa.adicionar_mensagem(MensagemVoz("audio.mp3"))
conversa.adicionar_mensagem(MensagemImagem("foto.jpg"))
conversa.exibir_todas()
````

### Duck Typing 🦆
Você conhece o "[Teste do pato](https://pt.wikipedia.org/wiki/Teste_do_pato)"?

> Se parece com um pato, nada como um pato e grasna como um pato, então provavelmente é um pato.

O teste diz que podemos reconhecer uma entidade desconhecida através de **comportamentos** familiares. Na orientação a objetos, a palavra "comportamento" mapeia diretamente para métodos de uma classe. No [glossário do Python](https://docs.python.org/3/glossary.html#term-duck-typing), duck typing é definido como (texto adaptado):

> *um estilo de programação que não verifica o tipo de um objeto; em vez disso, o método ou atributo é simplesmente chamado ou usado. (...) o código bem projetado melhora sua flexibilidade ao permitir a substituição polimórfica.*

A particularidade do duck typing é que **ele não precisa envolver herança, basta que dois objetos implementem o mesmo comportamento**. Já conhecemos isso através dos protocolos de uso dos métodos mágicos. Por exemplo, qualquer classe que implemente os métodos mágicos `__getitem__()` e `__setitem__()` podem ser utilizados como lista sem precisar herdar nenhum tipo.

Como Python não é uma linguagem fortemente tipada, contanto que um determinado tipo possua os atributos e métodos necessários para realizar um dado comportamento, seu tipo não importa. Vejamos o exemplo traduzido deste [tutorial do Real Python sobre duck typing](https://realpython.com/duck-typing-python/#duck-typing-behaving-like-a-duck).

[![image alt text](http://img.youtube.com/vi/Nn7PQwXP0kU/0.jpg)](http://www.youtube.com/watch?v=Nn7PQwXP0kU)



````python
class Pato:
    def nadar(self):
        print("O pato está nadando")

    def grasnar(self):
        print("O pato diz: AAAAAAAA")

class Marreco:
    def nadar(self):
        print("O marreco está nadando")

    def grasnar(self):
        print("O marreco diz: AAAAAAAA")

class Ganso:
    def nadar(self):
        print("O ganso está nadando")

    def grasnar(self):
        print("O ganso diz: AAAAAAAA")
````

````python
patos = [Pato(), Marreco(), Ganso()]

for pato in patos:
    pato.nadar()
    pato.grasnar()
    print('-------')
````
